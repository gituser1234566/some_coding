```python
import matplotlib.pyplot as plt
```

```python
epochs = range(1, 6)
```

```python
# Losses and F1 scores for each language
en_loss = {"name":"en", "data":[0.81, 0.34, 0.25, 0.22, 0.21]}
en_f1 = {"name":"en", "data":[0.08, 0.47, 0.45, 0.46, 0.63]}

de_loss = {"name": "de", "data": [0.74,0.30,0.23,0.204,0.202]}
de_f1 = {"name": "de", "data": [0.24,0.55,0.64,0.66,0.67]}

it_loss = {"name": "it", "data": [0.75,0.29,0.2223,0.199,0.188]}
it_f1 = {"name": "it", "data": [0.24,0.44,0.39,0.40,0.41]}

af_loss = {"name": "af", "data": [0.81,0.30,0.226,0.206,0.196]}
af_f1 = {"name": "af", "data": [0.16,0.58,0.6337,0.67,0.67]}

sw_loss = {"name": "sw", "data": [0.76,0.32,0.249,0.224,0.2176]}
sw_f1 = {"name": "sw", "data": [0.148,0.383,0.451,0.473,0.478]}
```

```python

all_losses = [en_loss, de_loss, it_loss, af_loss, sw_loss]
all_f1 = [en_f1, de_f1, it_f1, af_f1, sw_f1]
```

```python

colors = ['b', 'g', 'r', 'c', 'm']  # Different colors for each loss series
```

```python
plt.figure(figsize=(12, 8))

for i in range(len(all_losses)):
    plt.plot(epochs, all_losses[i]['data'], marker='o', linestyle='-', label=all_losses[i]['name'], color=colors[i])

plt.title('Electra Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.legend()
plt.grid(True)
plt.savefig('electra_loss.png')
plt.show()
```

```python
#clear figure
plt.clf()
plt.figure(figsize=(12, 8))

for i in range(len(all_f1)):
    plt.plot(epochs, all_f1[i]['data'], marker='o', linestyle='-', label=all_f1[i]['name'], color=colors[i])

plt.title('Electra macro-F1')
plt.xlabel('Epoch')
plt.ylabel('macro-F1 Score')
plt.xticks(epochs)
plt.legend()
plt.grid(True)
plt.savefig("electra_f1")
```
