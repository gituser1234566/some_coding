```python
import matplotlib.pyplot as plt
```

```python
epochs = range(1, 6)
```

```python
# Losses and F1 scores for each language
en_loss = {"name":"en", "data":[0.3422, 0.1433, 0.1207, 0.1071, 0.0990]}
en_f1 = {"name":"en", "data":[0.4622, 0.7915, 0.7988, 0.8029, 0.8112]}

de_loss = {"name": "de", "data": [0.2069, 0.07, 0.05, 0.04, 0.035]}
de_f1 = {"name": "de", "data": [0.55, 0.58, 0.59, 0.60, 0.605]}

it_loss = {"name": "it", "data": [0.2196, 0.0682, 0.0504, 0.0407, 0.0338]}
it_f1 = {"name": "it", "data": [0.6408, 0.6699, 0.6735, 0.6807, 0.6845]}

en_de_loss = {"name": "en_de", "data": [0.1726, 0.0943, 0.0733, 0.0586, 0.0478]}
en_de_f1 = {"name": "en_de", "data": [0.7651, 0.8043, 0.8184, 0.8245, 0.8297]}

en_it_loss = {"name": "en_it", "data": [0.3017, 0.1314, 0.1119, 0.0983, 0.0903]}
en_it_f1 = {"name": "en_it", "data": [0.5735, 0.4731, 0.6031, 0.806, 0.8093]}

en_it_de_loss = {"name": "en_it_de", "data": [0.206, 0.101, 0.0854, 0.0748, 0.0661]}
en_it_de_f1 = {"name": "en_it_de", "data": [0.7845, 0.7903, 0.8132, 0.8043, 0.8153]}
```

```python

all_losses = [en_loss, de_loss, it_loss, en_de_loss, en_it_loss, en_it_de_loss]
all_f1 = [en_f1, de_f1, it_f1, en_de_f1, en_it_f1, en_it_de_f1]
```

```python

colors = ['b', 'g', 'r', 'c', 'm', 'y']  # Different colors for each loss series
```

```python
plt.figure(figsize=(12, 8))

for i in range(len(all_losses)):
    plt.plot(epochs, all_losses[i]['data'], marker='o', linestyle='-', label=all_losses[i]['name'], color=colors[i])

plt.title('Comparison of Losses per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.legend()
plt.grid(True)
plt.savefig('losses.png')
plt.show()
```

```python
#clear figure
plt.clf()
plt.figure(figsize=(12, 8))

for i in range(len(all_f1)):
    plt.plot(epochs, all_f1[i]['data'], marker='o', linestyle='-', label=all_f1[i]['name'], color=colors[i])

plt.title('Comparison of macro-F1 per Epoch')
plt.xlabel('Epoch')
plt.ylabel('macro-F1 Score')
plt.xticks(epochs)
plt.legend()
plt.grid(True)
plt.savefig("f1")
```
